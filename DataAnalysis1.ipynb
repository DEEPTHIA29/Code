```python
import pandas as pd
import numpy as np

df=pd.read_csv("Data.csv")
df
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
impute=SimpleImputer(missing_values=np.nan,strategy='mean')
impute.fit(df.iloc[:,1:3])
x=impute.transform(df.iloc[:,1:3])
df.iloc[:,1:3]=x
df
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer.fit(df.iloc[:,:1].values)
df.iloc[:,:1]=imputer.transform(df.iloc[:,:1].values)
df
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer.fit(df.iloc[:,:].values)
df.iloc[:,:]=imputer.transform(df.iloc[:,:].values)
df
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer.fit(df.iloc[:,:3].values)
df.iloc[:,:3]=imputer.transform(df.iloc[:,:1].values)
df
```
