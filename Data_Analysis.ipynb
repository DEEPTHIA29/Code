```python
#FINDING THE AVERAGE RATING
df=data.dropna()#to drop the rows with null values
df.isnull()#to check how each values of columns whether it is null or not, if it is not null then it will display 'False' else it will display 'True'
df.isnull().sum()#Displays the number of null values in each columns
float(str(int(sum(df['Rating']))/9360)[:4])
```

```python
#FINDING THE NUMBER OF APPS HAVING RATING MORE THAN 5.0
df=data.dropna()
df.isnull().sum()
sum=0
for i in df['Rating']:
  if i==5.0:
    sum+=1
print(sum)
```

```python
#FINDING ABOUT THE NUMBER OF REVIEWS
s=0
for i in df['Reviews']:
  s+=int(i)
print(int(int(s)/(len(df['Reviews']))))
```

```python
#FINDING HOW MANY APPS ARE THERE BETWEEN RATING 4.0 TO 4.5
df=data.dropna()
df.isnull().sum()
sum=0
for i in df['Rating']:
  if (i>=4.0 and i<=4.5):
    sum+=1
print(sum)
```
